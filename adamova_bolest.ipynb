```r
data_1 <- read.csv("C:/Users/jsuch/Desktop/scholz/data - Skupina 1.csv") #nactu data
data_3 <- read.csv("C:/Users/jsuch/Desktop/scholz/data - Skupina 3.csv") #nactu data

# Function to convert columns from factor to numeric
convert_factor_to_numeric <- function(data) {
  data[, 6:ncol(data)] <- lapply(data[, 6:ncol(data)], function(column) {
    if (is.factor(column)) {
      as.numeric(as.character(column))
    } else {
      column
    }
  })
  return(data)
}

# Apply the function to data_1 and data_3
data_1 <- convert_factor_to_numeric(data_1)
data_3 <- convert_factor_to_numeric(data_3)
```

Adame, kód níže vezme pro skupiny 1 a 3 každý sloupec a použije Wilcoxonův test k otestování, jestli můžeme tvrdit, že jedna skupina má nižší nebo vyšší hodnotu pro daný sloupec než druhá. Kdyžtak zkonzultuj umělou inteligenci nebo někoho ze semináře nebo mě. Ale v první řadě zkus googlit a použít AI.

```r
for(i in 6:ncol(data_1)){
    mean_1 <- mean(data_1[,i])
    mean_3 <- mean(data_3[,i])
    print(colnames(data_1)[i])
    if(mean_1 < mean_3){
        test_result <- wilcox.test(data_1[,i], data_3[,i], alternative = "less")
        print("Prumer pro skupinu 1  je mensi nez prumer pro skupinu 3")
    } else {
        test_result <- wilcox.test(data_1[,i], data_3[,i], alternative = "greater")
        print("Prumer pro skupinu 1  je vetsi nez prumer pro skupinu 3")
    }
    
    print(test_result)
}

```

```r
# Function to calculate Pearson correlation for specific columns
calculate_pearson_correlations <- function(data, target_column, pattern) {
  # Ensure the target column exists
  if (ncol(data) < target_column) {
    stop("The target column index exceeds the number of columns in the dataframe.")
  }
  
  # Get column names matching the pattern
  matching_columns <- grep(pattern, colnames(data), value = TRUE)
  
  # Initialize a result list
  results <- list()
  
  # Iterate over matching columns
  for (col_name in matching_columns) {
    # Calculate Pearson correlation
    correlation <- cor(data[[target_column]], data[[col_name]], use = "complete.obs", method = "pearson")
    
    # Store the result
    results[[col_name]] <- correlation
  }
  
  return(results)
}

# Calculate Pearson correlations for the third column against columns with "CD21low" in their names
pearson_results_1 <- calculate_pearson_correlations(data_1, target_column = 3, pattern = "CD21low")
pearson_results_3 <- calculate_pearson_correlations(data_3, target_column = 3, pattern = "CD21low")

# Display results
print(pearson_results_1)

```
