```r
data_1 <- read.csv("C:/Users/jsuch/Desktop/scholz/data - Skupina 1.csv") #nactu data
data_3 <- read.csv("C:/Users/jsuch/Desktop/scholz/data - Skupina 3.csv") #nactu data

# Function to convert columns from factor to numeric
convert_factor_to_numeric <- function(data) {
  data[, 6:ncol(data)] <- lapply(data[, 6:ncol(data)], function(column) {
    if (is.factor(column)) {
      as.numeric(as.character(column))
    } else {
      column
    }
  })
  return(data)
}

# Apply the function to data_1 and data_3
data_1 <- convert_factor_to_numeric(data_1)
data_3 <- convert_factor_to_numeric(data_3)
```

Adame, kód níže vezme pro skupiny 1 a 3 každý sloupec a použije Wilcoxonův test k otestování, jestli můžeme tvrdit, že jedna skupina má nižší nebo vyšší hodnotu pro daný sloupec než druhá. Kdyžtak zkonzultuj umělou inteligenci nebo někoho ze semináře nebo mě. Ale v první řadě zkus googlit a použít AI.

```r
for(i in 6:ncol(data_1)){
    mean_1 <- mean(data_1[,i])
    mean_3 <- mean(data_3[,i])
    print(colnames(data_1)[i])
    if(mean_1 < mean_3){
        test_result <- wilcox.test(data_1[,i], data_3[,i], alternative = "less")
        print("Prumer pro skupinu 1  je mensi nez prumer pro skupinu 3")
    } else {
        test_result <- wilcox.test(data_1[,i], data_3[,i], alternative = "greater")
        print("Prumer pro skupinu 1  je vetsi nez prumer pro skupinu 3")
    }
    
    print(test_result)
}

```

```r
# Function to calculate Pearson correlation for specific columns
calculate_pearson_correlations <- function(data, target_column, pattern) {
  # Ensure the target column exists
  if (ncol(data) < target_column) {
    stop("The target column index exceeds the number of columns in the dataframe.")
  }
  
  # Get column names matching the pattern
  matching_columns <- grep(pattern, colnames(data), value = TRUE)
  
  print(matching_columns)
  # Initialize a result list
  results <- list()
  
  # Iterate over matching columns
  for (col_name in matching_columns) {
    
    # Calculate Pearson correlation
    correlation <- cor(data[[target_column]], data[[col_name]], use = "complete.obs", method = "pearson")
    
    # Linear regression
    model <- lm(data[[col_name]] ~ data[[target_column]], data = data)  
    model_summary <- summary(model)
    regression_coeff <- model_summary$coefficients[2, 1]  # Coefficient for independent variable
    p_value <- model_summary$coefficients[2, 4]    
    
    # Create the plot
    plot <- ggplot(data, aes(x = data[[target_column]], y = data[[col_name]])) +
      geom_point(color = "blue", alpha = 0.6) +
      geom_smooth(method = "lm", color = "red", se = FALSE) +
      labs(
        title = "",
        x = "",  # Adjusted: Use current column name for x-axis
        y = ""  # Adjusted: Use target column name for y-axis
      ) +
      theme_minimal() +
      annotate(
        "text",
        x = Inf, y = -Inf, hjust = 1.1, vjust = -0.5,
        label = paste0(
          "Pearsonova korelace: ", round(correlation, 2), "\n",
          "Regresní koeficient pro nezávislou proměnnou: ", round(regression_coeff, 2), "\n",
          "p-hodnota: ", format.pval(p_value, digits = 3)
        ),
        size = 4, color = "black"
      )
    
    # Print the plot
    print(plot)
    
    # Store the result
    results[[col_name]] <- correlation
  }
  
}

# Calculate Pearson correlations for the third column against columns with "CD21low" in their names
pearson_results_1 <- calculate_pearson_correlations(data_1, target_column = 3, pattern = "CD21low")

```

```r
# Load necessary libraries
library(ggplot2)
library(dplyr)

# Function to create comparative boxplots with points for all datasets
create_comparative_boxplots <- function(data_1, data_2, data_3) {
  # Get the column names for the sixth-to-last and subsequent columns
  column_names <- colnames(data_1)
  start_col <- ncol(data_1) - 5  # Sixth-to-last column index
  selected_columns <- column_names[start_col:ncol(data_1)]  # Select columns from 6th-to-last
  
  # Iterate over the selected columns
  for (col_name in selected_columns) {
    # Combine all three datasets into a long format
    combined_data <- bind_rows(
      data_1[, col_name, drop = FALSE] %>% mutate(Source = "Data 1"),
      data_2[, col_name, drop = FALSE] %>% mutate(Source = "Data 2"),
      data_3[, col_name, drop = FALSE] %>% mutate(Source = "Data 3")
    )
    combined_data <- combined_data %>% rename(Value = !!col_name)
    
    # Create the boxplot with points for each dataset
    plot <- ggplot(combined_data, aes(x = Source, y = Value, fill = Source)) +
      geom_boxplot(outlier.color = "red", alpha = 0.5) +  # Boxplots for all datasets
      geom_point(aes(color = Source), size = 2, position = position_jitter(width = 0.2)) +  # Points for all datasets
      scale_fill_manual(values = c("Data 1" = "lightblue", "Data 2" = "lightgreen", "Data 3" = "orange")) +  # Custom fill colors
      scale_color_manual(values = c("Data 1" = "blue", "Data 2" = "green", "Data 3" = "orange")) +  # Custom point colors
      labs(
        title = paste("Boxplot for", col_name),
        x = "Data Source",
        y = col_name
      ) +
      theme_minimal()
    
    # Print the plot
    print(plot)
  }
}

# Call the function for the three datasets
create_comparative_boxplots(data_1, data_2, data_3)
```
